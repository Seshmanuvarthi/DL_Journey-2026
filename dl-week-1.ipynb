```python
import torch
import torch.nn as nn
import torch.optim as optim

x = torch.randn(10, 100)
y = torch.randn(10, 1)


class SimpleNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(100, 1)  #Creates a linear layer with 100 inputs and 1 output

    def forward(self,x):
        return self.linear(x)

model=SimpleNetwork()

loss_fn = nn.MSELoss()   
optimiser=optim.Adam(model.parameters(),lr=0.01)

for epoch in range(100):            
    y_pred = model(x)                 
    loss = loss_fn(y_pred, y)         

    optimiser.zero_grad()
    loss.backward()
    optimiser.step()
                 

print(f"Final loss:", loss.item())   
    
```

```python
import tensorflow as tf                    
x = tf.random.normal((100, 3))          
y = tf.random.normal((100, 1))            

# Define model
model = tf.keras.Sequential([             
    tf.keras.layers.Dense(1,               
                          input_shape=(3,))
])

model.compile(optimizer='adam',            
              loss='mse')                  

# Train model
model.fit(x, y,                             
          epochs=100,                      
          verbose=0)                       

print("Final loss:",                      
      model.evaluate(x, y))               
```

```python
#Basic perceptron 
import numpy as np
def step(val):
    if val>=0:
        return 1
    else:
        return 0

class Basic_Perceptron:

    def __init__(self,weights,bais):
        self.weights=weights
        self.bais=bais

    def predict(self,inputs):
        op=np.dot(weights,inputs)+bais
        return step(op)

weights = np.array([1, 1])            
bais = -1.5  

model=Basic_Perceptron(weights,bais)
print("AND Gate")
for x in [(0,0), (0,1), (1,0), (1,1)]:
    print(x, "->", model.predict(np.array(x)))

bais=-0.5
model2=Basic_Perceptron(weights,bais)
                    
print("\nOR Gate")
for x in [(0,0), (0,1), (1,0), (1,1)]:
    print(x, "->", model2.predict(np.array(x)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist

print("Loading MNIST Data...")
(x_train, _), (x_test, _) = mnist.load_data()

# Normalize pixel values to be between 0 and 1
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Reshape the images to include the color channel dimension (28x28x1)
x_train = np.expand_dims(x_train, axis=-1)
x_test = np.expand_dims(x_test, axis=-1)

print(f"Training data shape: {x_train.shape}")
print(f"Testing data shape: {x_test.shape}")

input_img = layers.Input(shape=(28, 28, 1))

#encoder
# Compress the 28x28 image down to a 7x7 representation
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x) # Shrinks to 14x14

x = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(x)
encoded = layers.MaxPooling2D((2, 2), padding='same')(x) # Shrinks to 7x7

# decoder
# Rebuild the 7x7 representation back up to 28x28
x = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(encoded)
x = layers.UpSampling2D((2, 2))(x) # Grows to 14x14

x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x) # Grows to 28x28

# Output layer uses sigmoid to ensure pixel values stay exactly between 0 and 1
decoded = layers.Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = models.Model(input_img, decoded)

# 'binary_crossentropy' works exceptionally well here because our pixels are scaled 0 to 1
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.summary()

print("\nStarting Autoencoder Training...")
# Autoencoders learn by trying to make the output match the input (x_train, x_train)
history = autoencoder.fit(
    x_train, 
    x_train, 
    epochs=10, # 10 epochs is a good sweet spot for MNIST 
    batch_size=256, 
    shuffle=True, 
    validation_data=(x_test, x_test)
)


#Visualisation 
print("\nGenerating visualizations...")
decoded_imgs = autoencoder.predict(x_test)

n = 10  # How many digits we will display
plt.figure(figsize=(20, 4))
for i in range(n):
    # Display the original digit
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(x_test[i].reshape(28, 28), cmap='gray')
    plt.title("Original")
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display the reconstructed digit
    ax = plt.subplot(2, n, i + 1 + n)
    plt.imshow(decoded_imgs[i].reshape(28, 28), cmap='gray')
    plt.title("Reconstructed")
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np

data_dir = '/kaggle/input/datasets/aneesh10/cricket-shot-dataset/data/'

# We use 64x64 to keep memory usage safe while proving the concept
img_size = 64 
input_dim = img_size * img_size * 3 # 64 * 64 * 3 = 12,288 dimensions

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir, validation_split=0.2, subset="training", seed=123,
    image_size=(img_size, img_size), batch_size=32, label_mode=None 
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir, validation_split=0.2, subset="validation", seed=123,
    image_size=(img_size, img_size), batch_size=32, label_mode=None
)

# Normalize to [0, 1] and format as (Input, Target)
def prepare_data(img):
    img = img / 255.0
    return img, img

train_ds = train_ds.map(prepare_data)
val_ds = val_ds.map(prepare_data)

input_img = layers.Input(shape=(img_size, img_size, 3))

# ENCODER: Shrink spatial size using MaxPooling
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x) # Shrinks to 32x32

x = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(x)

encoded_under = layers.MaxPooling2D((2, 2), padding='same')(x) 

# DECODER: Rebuild the image
x = layers.Conv2D(8, (3, 3), activation='relu', padding='same')(encoded_under)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
decoded_under = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

under_ae = models.Model(input_img, decoded_under)
under_ae.compile(optimizer='adam', loss='mse')

# ENCODER: Instead of pooling (shrinking), we INCREASE the filters
# BOTTLENECK: 64x64 pixels with 16 filters = 65,536 dimensions (Massive Expansion!)
encoded_over = layers.Conv2D(16, (3, 3), activation='relu', padding='same')(input_img)

# DECODER: Map it back down to 3 channels (RGB)
decoded_over = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(encoded_over)

over_ae = models.Model(input_img, decoded_over)
over_ae.compile(optimizer='adam', loss='mse')

print(f"Original Image Dimensions: {input_dim}")
print("Training Undercomplete (12,288 -> 2,048 -> 12,288)...")
under_ae.fit(train_ds, epochs=15, validation_data=val_ds, verbose=1)

print("\nTraining Overcomplete (12,288 -> 65,536 -> 12,288)...")
over_ae.fit(train_ds, epochs=15, validation_data=val_ds, verbose=1)
print("\nGenerating visualizations...")
for test_images, _ in val_ds.take(1):
    under_reconstructions = under_ae.predict(test_images)
    over_reconstructions = over_ae.predict(test_images)
    
    n = min(5, test_images.shape[0])
    plt.figure(figsize=(15, 8))
    
    for i in range(n):
        # Original
        ax = plt.subplot(3, n, i + 1)
        plt.imshow(test_images[i].numpy())
        plt.title("Original\n(12,288 dim)")
        plt.axis('off')
        
        # Undercomplete
        ax = plt.subplot(3, n, i + 1 + n)
        plt.imshow(under_reconstructions[i])
        plt.title("Undercomplete\n(2,048 dim)")
        plt.axis('off')
        
        # Overcomplete
        ax = plt.subplot(3, n, i + 1 + 2*n)
        plt.imshow(over_reconstructions[i])
        plt.title("Overcomplete\n(65,536 dim)")
        plt.axis('off')
        
    plt.tight_layout()
    plt.show()
```

The "Half-Pipe" Compression
If you do the math on the bottleneck, the original image is 28x28 (784 pixels). After passing through the two MaxPooling layers, the image is shrunk down to a 7x7 grid with 8 filters. That leaves us with 392 values in the latent space. So, the network is compressing the image by exactly 50%. It's an undercomplete autoencoder, but a pretty gentle one. Because it's only dropping half the data, it doesn't have to sweat too hard to rebuild the numbers, which is why the output looks so good.

2. Why it beats the Dense version
By using Conv2D instead of Dense layers, we aren't forcing the image into a single flat line of pixels. The network gets to keep the 2D grid shape intact. This means it naturally picks up on spatial features—it learns what a "loop" looks like for an 8, or what a "diagonal line" looks like for a 7, rather than just memorizing pixel coordinates.

3. The weird (but clever) Loss Function
Normally, you'd expect to see Mean Squared Error (MSE) used to compare two images. But here, the code uses binary_crossentropy. This usually belongs in classification tasks, but it works perfectly here because we divided all the pixels by 255.0. Since every pixel is now a value between 0 and 1, the network treats the pixel intensity like a probability (e.g., "I am 90% sure this pixel should be white").

4. The "Photoshop Smoothing" Effect
When you look at the plotted results, the reconstructed digits in the bottom row usually look a little softer or blurrier than the top row. It acts like an accidental noise filter. If a person had really shaky handwriting in the original test image, the autoencoder tends to smooth those jagged edges out. It’s not copy-pasting the image; it’s drawing what it thinks the "ideal" version of that shape should look like based on the features it saved in the bottleneck.

```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
import matplotlib.pyplot as plt
import numpy as np

data_dir = '/kaggle/input/datasets/aneesh10/cricket-shot-dataset/data/'
img_size = 64 

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir, validation_split=0.2, subset="training", seed=123,
    image_size=(img_size, img_size), batch_size=32, label_mode=None 
)
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir, validation_split=0.2, subset="validation", seed=123,
    image_size=(img_size, img_size), batch_size=32, label_mode=None
)

def prepare_data(img):
    img = img / 255.0
    return img, img

train_ds = train_ds.map(prepare_data)
val_ds = val_ds.map(prepare_data)

input_img = layers.Input(shape=(img_size, img_size, 3))

# ENCODER (Overcomplete + Regularized)
# We use 32 filters (which is a massive expansion, 131,072 dimensions!)
# BUT we add L1 Regularization. This is the magic bullet.
encoded_sparse = layers.Conv2D(
    filters=32, 
    kernel_size=(3, 3), 
    activation='relu', 
    padding='same',
    activity_regularizer=regularizers.l1(1e-6) # The Sparsity Penalty
)(input_img)

# DECODER
decoded_sparse = layers.Conv2D(
    filters=3, 
    kernel_size=(3, 3), 
    activation='sigmoid', 
    padding='same'
)(encoded_sparse)

sparse_ae = models.Model(input_img, decoded_sparse)
sparse_ae.compile(optimizer='adam', loss='mse')

print("Training Sparse (Regularized) Autoencoder...")
# We use more epochs because regularization makes it harder for the network to learn
sparse_ae.fit(train_ds, epochs=25, validation_data=val_ds, verbose=1)
print("\nGenerating visualizations...")
for test_images, _ in val_ds.take(1):
    sparse_reconstructions = sparse_ae.predict(test_images)
    
    n = min(5, test_images.shape[0])
    plt.figure(figsize=(15, 6))
    
    for i in range(n):
        # Original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(test_images[i].numpy())
        plt.title("Original")
        plt.axis('off')
        
        # Sparse Reconstruction
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(sparse_reconstructions[i])
        plt.title("Sparse\n(L1 Regularized)")
        plt.axis('off')
        
    plt.tight_layout()
    plt.show()
```
